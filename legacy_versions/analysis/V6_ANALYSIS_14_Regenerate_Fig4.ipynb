```python
# ==============================================================================
# NOTEBOOK FILENAME: V6_ANALYSIS_14_Regenerate_Fig4.ipynb
#
# DESCRIPTION:
# This utility script corrects a critical typo in the legend of the original
# Figure 3 (now Figure 4 in the manuscript). It re-plots the data from
# 'Open_Loop_Fsig_vs_T_1D.csv' with the corrected legend label
# ("Response Amplitude" instead of "Driving Amplitude").
#
# AUTHOR: Gemini (as the next-generation AI assistant)
# DATE: 2025-08-22
# ==============================================================================

# --- Part 1: Setup and Library Imports ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

print("Setup complete. Libraries imported.")

# --- Part 2: Upload the source data file ---
print("\nPlease upload the data file for the plot:")
print("--> Open_Loop_Fsig_vs_T_1D.csv <--")

try:
    uploaded = files.upload()
    csv_filename = list(uploaded.keys())[0]
    print(f"Successfully uploaded '{csv_filename}'.")
    df = pd.read_csv(csv_filename)
except (ValueError, IndexError, FileNotFoundError):
    print("\n[ERROR] File upload was cancelled or failed. Please run the cell again to retry.")
    raise SystemExit

# --- Part 3: Plotting the Corrected Figure ---
print("Generating the corrected plot...")

# --- Plotting settings (to match the original figure style) ---
plt.style.use('default')
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Liberation Serif']
plt.rcParams['mathtext.fontset'] = 'dejavuserif'
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
plt.rcParams['legend.fontsize'] = 10
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.linestyle'] = ':'

# --- Create the plot ---
fig, ax = plt.subplots(figsize=(7, 5))

# Plot the data with error bars
ax.errorbar(df['T'], df['mean_F_sig0'], yerr=df['sem_F_sig0'],
            fmt='o-', capsize=4, color='red', markersize=4, linewidth=1.5,
            # --- CRITICAL CORRECTION HERE ---
            label=r'Mean Response $F_{\mathrm{sig},0}$ ( SEM, 1D Model)')

# Add the Kendall's tau annotation
ax.text(0.95, 0.95, "Kendall's $\\tau = -0.64, p < 0.001$",
        transform=ax.transAxes, fontsize=10,
        verticalalignment='top', horizontalalignment='right',
        bbox=dict(boxstyle='round,pad=0.5', fc='wheat', alpha=0.5))

# Set labels and limits
ax.set_xlabel('Thermal Strength (T) (dimensionless)')
ax.set_ylabel(r'Open-Loop Response Amplitude $F_{\mathrm{sig},0}$ (dimensionless)')
ax.set_xlim(0, 2.1)
ax.set_ylim(bottom=0)
ax.ticklabel_format(style='sci', axis='y', scilimits=(0,0)) # Scientific notation for y-axis

ax.legend()
plt.tight_layout()

# --- Save the corrected figure ---
output_filename = "fig3_open_loop_response_corrected.pdf"
plt.savefig(output_filename, format='pdf', dpi=300, bbox_inches='tight')

print(f"Generated corrected plot file: {output_filename}")
plt.show()

# --- Part 4: Automatic Download ---
print(f"\nPreparing to download the corrected figure...")
files.download(output_filename)

print("\n--- FIGURE 4 REGENERATION TASK COMPLETE ---")
```

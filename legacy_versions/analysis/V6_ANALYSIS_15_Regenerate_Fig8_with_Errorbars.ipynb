```python
# ==============================================================================
# NOTEBOOK FILENAME: V6_ANALYSIS_15_Regenerate_Fig8_with_Errorbars.ipynb
#
# DESCRIPTION:
# This utility script resolves the first part of Task D-8. It regenerates
# Figure 8 (the harmonic ratio plot) to include error bars (SEM),
# enhancing its scientific robustness as per the reviewer's suggestion.
# It loads the existing data and outputs a new, final PDF for the figure.
#
# AUTHOR: Gemini (as the next-generation AI assistant)
# DATE: 2025-08-22
# ==============================================================================

# --- Part 1: Setup and Library Imports ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
import os

print("Setup complete. Libraries imported.")

# --- Part 2: Upload the source data file ---
# This script uses the data generated in the Task B simulation.
print("\nPlease upload the data file for the plot:")
print("--> Harmonic_Ratio_vs_T.csv <--")

try:
    uploaded = files.upload()
    csv_filename = list(uploaded.keys())[0]
    print(f"Successfully uploaded '{csv_filename}'.")
    df = pd.read_csv(csv_filename)
except (ValueError, IndexError, FileNotFoundError):
    print("\n[ERROR] File upload was cancelled or failed. Please run the cell again to retry.")
    raise SystemExit

# --- Part 3: Plotting the Corrected Figure with Error Bars ---
print("Generating the plot with error bars...")

# --- Plotting settings (to match the original figure style) ---
plt.style.use('default')
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Liberation Serif']
plt.rcParams['mathtext.fontset'] = 'dejavuserif'
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
plt.rcParams['legend.fontsize'] = 10
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.linestyle'] = ':'

# --- Create the plot ---
fig, ax = plt.subplots(figsize=(7, 5))

# --- MODIFIED: Use ax.errorbar to include SEM for the ratio ---
ax.errorbar(df['T'], df['ratio_3w_1w'], yerr=df['sem_ratio'],
            fmt='o-', capsize=4, color='purple', markersize=5, linewidth=1.5,
            label='3rd Harmonic / Fundamental Ratio')

# Set labels and limits
ax.set_xlabel('Thermal Strength (T) (dimensionless)')
ax.set_ylabel('Harmonic Ratio ($A_{3\omega} / A_{1\omega}$)')
ax.set_xlim(0, 2.1)
ax.set_ylim(bottom=0)
ax.set_title('Nonlinearity vs. Noise (at $A_{drive} = 0.001$)')
ax.legend()

plt.tight_layout()

# --- Save the corrected figure with a new, final filename ---
output_filename = "v6_fig08_harmonic_ratio_with_errors.pdf"
plt.savefig(output_filename, format='pdf', dpi=300, bbox_inches='tight')

print(f"Generated new plot file with error bars: {output_filename}")
plt.show()

# --- Part 4: Automatic Download ---
print(f"\nPreparing to download the new figure...")
files.download(output_filename)

print("\n--- FIGURE 8 REGENERATION TASK COMPLETE ---")
```
